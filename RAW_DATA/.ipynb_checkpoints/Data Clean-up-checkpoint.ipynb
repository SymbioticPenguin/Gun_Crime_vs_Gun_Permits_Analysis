```python
import pandas as pd
import numpy as np

file = "./gun-violence-data_01-2013_03-2018.csv"
file2 = './nics-firearm-background-checks.csv'

violence = pd.read_csv(file)
permit = pd.read_csv(file2)
```

```python
#violence.count()
```

```python
violence_new = violence[['incident_id','date','state','city_or_county','latitude','longitude']]
```

```python
permit["year"] = permit['month'].str.split('-').str[0]

permit['year'] = permit['year'].astype('int32')

permit_new = permit.loc[permit['year']>=2013]
# permit
```

```python
permit_new = permit_new.filter(['year','state','totals'])
permit_new
```

```python
data = {"year":np.arange(2013,2019)}

permit_sum = pd.DataFrame(data)
permit_sum
total_permits = []                        
for i in range(2013,2019):
     total_permits.append(permit_new.loc[permit_new['year']==i]['totals'].sum())
permit_sum['total_permits'] = total_permits

permit_sum
```

```python
violence_new
# violence_new = violence_new.set_index('state')
# violence_new = violence_new.set_index(['state','city_or_county'])

violence_new = violence_new.dropna()
violence_new = violence_new.reset_index()
```

```python
years = []
for index,row in violence_new.iterrows():
    years.append(row['date'].split("-")[0])
```

```python
violence_new['date'] = years
violence_new['date'] = violence_new['date'].astype(int)
violence_new.head()
```

```python
violence_new.to_csv(r'crime_data_clean.csv', index = False)
permit_new.to_csv(r'permit_data_clean.csv',index = False)
permit_sum.to_csv(r'total_permits_by_year.csv',index = False)
```
